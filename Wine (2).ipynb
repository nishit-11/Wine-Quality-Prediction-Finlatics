Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Loading The Dataset

```python
df = pd.read_csv('wine_data.csv')
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

print (x)
print (y)
```

Finding the most frequently occuring, minimum and maximum quality.

```python
most = df['quality'].mode()[0]
mini = df['quality'].min()
maxi = df['quality'].max()

print("Most Frrequently Occuring Wine Quality is:", most)
print("Minimum Wine Quality is:", mini)
print("Maximum Wine Quality is:", maxi)
```

Finding how 'fixed acidity', 'alcohol content' and 'free sulphur dioxide' is co-related to quality of the wine

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[["alcohol", "fixed acidity", "free sulfur dioxide","quality"]].corr()
print(corr["quality"])

sns.heatmap(df[["alcohol", "fixed acidity", "free sulfur dioxide","quality"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Finding out average residual sugar in best quality and worst quality wine.

```python
best = df[df['quality'] == maxi]['residual sugar'].mean()
worst = df[df['quality'] == mini]['residual sugar'].mean()

print("Average Residual Sugar in Best Quality Wine is:", best)
print("Average Residual Sugar in Worst Quality Wine is:", worst)
```

Affect of Volatile Acidity on Quality of the wine.

```python
vol = df["volatile acidity"].corr(df["quality"])
print("Correlation between Volatile Acidity and Quality is:",vol)
```

Splitting the Dataset into Training and Testing data and scaling it.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
print(x_train_scaled)
print(x_test_scaled)
```

Training and Testing a Decision Tree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt.fit(x_train_scaled,y_train)
y_dt = dt.predict(x_test_scaled)

result_dt = pd.DataFrame({'Actual': y_test, 'Predicted': y_dt})
print(result_dt)
```

Training and Testing a Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train_scaled, y_train)
y_rf = rf.predict(x_test_scaled)

result_rf = pd.DataFrame({'Actual': y_test, 'Predicted': y_dt})
print(result_rf)
```

Comparing the accuracy of both the model.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score

acc_dt = accuracy_score(y_test, y_dt)
acc_rf = accuracy_score(y_test, y_rf)
r2_dt = r2_score(y_test, y_dt)
r2_rf = r2_score(y_test, y_rf)

compare = pd.DataFrame(
    {'Model': ['Decision Tree', 'Random Forest'],
    'Accuracy': [acc_dt, acc_rf],
     'R2 Score' : [r2_dt, r2_rf]})
print(compare)

cm = confusion_matrix(y_test, y_rf)
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
