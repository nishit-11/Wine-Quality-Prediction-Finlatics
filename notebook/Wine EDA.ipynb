Importing All the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('wine_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
most = df['quality'].mode()[0]
mini = df['quality'].min()
maxi = df['quality'].max()

print("Most Frrequently Occuring Wine Quality is:", most)
print("Minimum Wine Quality is:", mini)
print("Maximum Wine Quality is:", maxi)
```

```python
best = df[df['quality'] == maxi]['residual sugar'].mean()
worst = df[df['quality'] == mini]['residual sugar'].mean()

print("Average Residual Sugar in Best Quality Wine is:", best)
print("Average Residual Sugar in Worst Quality Wine is:", worst)
```

```python
vol = df["volatile acidity"].corr(df["quality"])
print("Correlation between Volatile Acidity and Quality is:",vol)
```

```python
corr = df[["alcohol", "fixed acidity", "free sulfur dioxide","quality"]].corr()
print(corr["quality"])

sns.heatmap(df[["alcohol", "fixed acidity", "free sulfur dioxide","quality"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
